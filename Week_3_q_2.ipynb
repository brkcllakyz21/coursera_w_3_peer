```python
import pandas as pd
```

    As you remember from first notebook we have created the pure data of toronto. Now we will add the latitude and longitude to the first dataframe which is toronto_pure_data. We are starting by reading the csv file which has latitude and longitude of the postal codes and named as toronto_geo_data

```python
toronto_geo_data = pd.read_csv("toronto_geo.csv")
toronto_geo_data
```

Now we have a merge operation at the junk below. We are combining both toronto_pure_data and toronto_geo_data dataframes on toronto_final_data by using merge method. The "on" statement at the merge method represents the column which we are combining data on. In our case this column is "PostalCode" column. Method matches the rows which have equal "PostalCode" values.

```python
toronto_final_data = toronto_pure_data.merge(toronto_geo_data, on="PostalCode", how = 'inner')
```
